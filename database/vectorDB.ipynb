```python
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams

def launchDB():
    # Initialize the client
    client = QdrantClient("localhost", port=6333)
    # Creat a collection
    client.create_collection(
        collection_name="lyudao_collection",
        vectors_config=VectorParams(size=1536, distance=Distance.DOT),
        )
    return client
```

```python
import json
from qdrant_client.http.models import PointStruct

def insertPoint(c, chunk):
    payload = {"id": chunk['id'],
                "conversation": chunk['conversation']} 
    return PointStruct(id=c, vector=chunk['embedding'], payload=payload)

def addVectors(client, points):
    operation_info = client.upsert(
        collection_name="lyudao_collection",
        wait=True,
        points=points
        )
    print(operation_info)

def SearchByVector(client, vector, limit):
    # run a query
    return client.search(
        collection_name="lyudao_collection", query_vector=vector, limit=limit
        )

def SearchByIDs(client,ids):
    return client.retrieve(
        collection_name="lyudao_collection",
        ids=ids,
        )

def main():
    client = launchDB()
    # Opening JSON file
    f = open('all_chunks.json')
    data = json.load(f)
    # insert points
    points = []
    for c in range(len(data)):
        points.append(insertPoint(c, data[c]))
    # Closing file
    f.close()
    
    # Add Vectors
    addVectors(client, points)
    # run queries
    print(SearchByVector(client, data[0]['embedding'], 3))
    print(SearchByIDs(client, [1]))

if __name__ == "__main__":
    main()
```
